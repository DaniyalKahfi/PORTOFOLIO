# IV. EXPLORATORY DATA ANALYSIS

### 4.1 Ketidakpuasan Pelanggan Berdasarkan Tipe Kelas Layanan

```python
# Membuat bar plot
plt.figure(figsize=(10, 6))
sns.countplot(x='class_type', hue='satisfaction', data=df)

# Memberikan label pada sumbu x
plt.xlabel('Class Type')

# Memberikan label pada sumbu y
plt.ylabel('Count')

# Menampilkan plot
plt.show()
```

Terlihat bahwa customer yang tidak puas terhadap layanan maskapai Invicto pada kategori kelas eco dan customer yang puas pada kelas business. <br>
<br>
Namun, jika dilihat perbandingan puas dan tidak puas pada setiap kelasnya, customer eco dan eco plus lebih banyak ketidakpuasan, berdasarkan referensi hal ini menjadi fenomena yang berbahaya terhadap perusahaan. 

```python

# Menghitung jumlah kepuasan dan ketidakpuasan
satisfaction_counts = df['satisfaction'].value_counts()

# Membuat pie chart
plt.figure(figsize=(8, 8))
plt.pie(satisfaction_counts, labels=satisfaction_counts.index, autopct='%1.1f%%', startangle=90)

# Menambahkan judul
plt.title('Customer Satisfaction Distribution')

# Menampilkan pie chart
plt.show()
```

Dari keseluruhan pelanggan yang menggunakan Invicto Airline, dapat disimpulkan bahwa 54.7% pelanggan puas terhadap layanan yang diberikan Invicto Airline. Sedangkan, 45,3% tidak puas. Ketidakpuasan pelanggan ini dianggap cukup tinggi. Dalam buku "Harvard Business Review Enterpreneur's Handbook"(2017), dijelaskan bahwa batas maksimum ketidakpuasan pelanggan yang dianggap wajar atau dapat diterima hanya berkisar 5-10%

```python
# Menghitung jumlah data untuk setiap nilai unik dalam kolom 'class_type'
class_type_counts = df['class_type'].value_counts()

# Membuat bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=class_type_counts.index, y=class_type_counts.values)

# Memberikan judul
plt.title('Distribution of Class Types')

# Memberikan label pada sumbu x
plt.xlabel('Class Type')
```
